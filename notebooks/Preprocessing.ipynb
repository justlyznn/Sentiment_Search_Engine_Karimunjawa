```python
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
df = pd.read_csv("../data/cleaning/clear_data.csv")
df.head()
```

#Normalisasi

```python
norm = {
    'bnr': 'benar', 'yg': 'yang', 'thn': 'tahun', 'wk': ' ', 'bpk': 'bapak',
    'lg': 'lagi', 'kyk': 'kayak', 'mkan': 'makan', 'bljr': 'belajar', 'mikir2': 'mikir',
    'siaaapp': 'siap', 'okaayy': 'oke', 'org': 'orang', 'one': 'satu', 'dg': 'dengan',
    'lali': 'lupa', 'pntai': 'pantai', 'htm': 'biaya', 'enk': 'enak', 'mntap': 'mantap', 'g': 'gak', 'cantikkkk ': 'cantik', 'buaguussss': 'bagus', 'bgttttt': 'banget', 'mjd': 'menjadi'
}

def normalisasi(str_text):
    if not isinstance(str_text, str):
        return str(str_text)
    for i in norm:
        str_text = re.sub(r'\b' + re.escape(i) + r'\b', norm[i], str_text)
    return str_text

df['text'] = df['text'].apply(lambda x: normalisasi(x))
df
```

#Stopwords

```python
import Sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary

more_stop_words = []

stop_words = StopWordRemoverFactory().get_stop_words()
stop_words.extend(more_stop_words)

new_array = ArrayDictionary(stop_words)
stop_words_remover_new = StopWordRemover(new_array)

def stopword(str_text):
    str_text = stop_words_remover_new.remove(str_text)
    return str_text

df['text'] = df['text'].apply(lambda x: stopword(x))
df.head()
```

#Tokenize

```python
tokenized = df['text'].apply(lambda x:x.split())
tokenized.head()
```

#Stemming

```python

print("DataFrame 'df' has been saved to '../data/preprocessing/preprocessing_data.csv'")
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(text_cleaning):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    do = []
    for w in text_cleaning:
        dt = stemmer.stem(w)
        do.append(dt)
    d_clean = " ".join(do)
    print(d_clean)
    return d_clean

tokenized = tokenized.apply(stemming)
tokenized.to_csv('../data/preprocessing/preprocessing_data.csv', index=False)
```
